### Question 1: Check if a number is Palindrome* [10%]

    *A palindrome number is a number that remains the same when its digits are reversed. In other words, it reads the same backward as forward. Palindrome numbers are symmetric and exhibit the same sequence of digits when read from left to right and from right to left.

<b>a. Write a code snippet for user to input a number. [2%]</b>

<b>b. Run a check to see if the input is a number or a string, if user input a string, display out a message showing, "Incorrect format, please input a number." [4%]</b>

<b>c. Once the user inputs a number, run a check to see if the number is palindrome or not. If the number is palindrome display "The given number is Palindrome". If it's not a palindrome, then display, "The given number is not a Palindrome. [4%]</b>

##### Examples:

input: "bib"<br>
output: Incorrect format, please input a number.

input: "415"<br>
output: The given number is not a Palindrome.

input: "212"<br>
output: The given number is Palindrome.




```python
def is_palindrome(num):
    str_num = str(num)
    
    return str_num == str_num[::-1]

# Part (a): User input
user_input = input("Enter a number: ")

# Part (b): Check if the input is a number or string
if user_input.isdigit():
    # Part (c): Check if the number is palindrome
    if is_palindrome(int(user_input)):
        print("The given number is Palindrome.")
    else:
        print("The given number is not a Palindrome.")
else:
    print("Incorrect format, please input a number.")
```

### Question 2:  Remove STOP words* [10%]

    *In NLP, stop words are commonly used words like "a", "is", and "the". 
    They are typically filtered out during processing.

<b>a. Implement a function that takes a string text and an integer k. [2%]</b>

<b>b. Checks the occurrence of each word in the sentence. [3%]</b>

<b>c. Returns the list of words that occur in the text at least k times. [3%]</b>

<b>d. The words must be returned in the order of their first occurrence in the text. [2%]</b>

##### Examples:

input:

    text = "a mouse is smaller than a dog but a dog is stronger"
    k=2

output: 

    ['a', 'is', 'dog']

##### Explanation: 
- The list of stop words that occur at least k = 2Â times is ["a", "is", "dog"]. 
- "a" occurs 3 times, "is" and "dog" both occur 2 times. No other word occurs at least 2 times. 
- The answer is in order of the first appearance in text.




```python
# Question 2: insert your code here
def stop(sentence, k):
    sentence = sentence.lower()
    words = sentence.split()
    words_count = {}  #create an empty dictionary to capture the count of each word
    for word in words:
        clean_word = ''.join(c for c in word if c.isalpha())  #This checks if the character c is an alphabetical character (a letter).
        if clean_word:
            words_count[clean_word] = words_count.get(clean_word,0) +1  #increment the occurence by 1 for each occurence
    word_list = []   
    for word,count in words_count.items():
        if count >= k:
            word_list.append(word)
    print(word_list)  
    
#user-input
sentence = input('Please input your sentence: ')
k = int(input('Please input an int for k: '))
stop(sentence,k)
   
```

### Question 3: Suffix stripping Stemmer* [10%]

    *Stemming is the process of extracting the base word from a word. For instance, the base for "worked" is "work".
    
<b>a) Implement a function that takes a sentence without punctuation. [1%]</b>

<b>b) If the word ends in 'ed', 'ly' or 'ing', remove the suffix. [4%]</b>

<b>c) If the resulting word is longer than 7 letters, keep the first 7 letters. [3%]</b>

<b>d) Return the stemmed counterpart of individual words. [2%]</b>

##### Example:

input:

    text = "an extremely ferocious dog is barking"

output: 

    an extreme ferocio dog is bark

##### Explanation: 
- 'an' does not end in one of the suffixes and is less than 8 letters
- 'extremely' is 'extreme' after removing the suffix and 'extreme' is less than 8 letters
- 'ferocious' is 9 letters long, so reduce it to 7 letters
- ' ferocio','dog' and 'is' are unchanged
- 'barking' is 'bark' after removing the suffix and is less than 7 letters

# usually to find pattern is to use Ragex

```python
import re

def stemmer(text):

    #remove the punctuation in the sentence
    clean_text = re.sub(r"([!,.?:;'])", '', text)  #sub = substitute these raw string with these punctuation marks with a blank
    #split the sentence into indiviual words
    words = clean_text.split()
    suffixes = ('ed','ly','ing')
    #create a list to store the individual words
    clean_words =[]
    for word in words:
        for suffix in suffixes:
            if word.endswith(suffix):
                word = word[:-len(suffix)]
                break
        if len(word) > 7:
            word = word[:7]
        clean_words.append(word)
    return ' '.join(clean_words)     #combine the words in the list to form back the sentence
    
#user-input
text = input('Please input your sentence: ')              
print(stemmer(text))  
```

#### We can use get_dummies from pandas function
pandas.get_dummies() is used for data manipulation. It converts categorical data into dummy or indicator variables.

syntax:  pandas.get_dummies(data, prefix=None, prefix_sep=â_â, dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)

Parameters:

data: whose data is to be manipulated.
<br>
prefix: String to append DataFrame column names. Pass a list with length equal to the number of columns when calling get_dummies on a DataFrame. Default value is None.
<br>
prefix_sep: Separator/delimiter to use if appending any prefix. Default is â_â
<br>
dummy_na: It adds a column to indicate NaN values, default value is false, If false NaNs are ignored.
<br>
columns: Column names in the DataFrame that needs to be encoded. Default value is None, If columns is None then all the columns with object or category dtype will be converted.
<br>
sparse: It  specify whether the dummy-encoded columns should be backed by a SparseArray (True) or a regular NumPy array (False). default value is False.
<br>
drop_first: Remove first level to get k-1 dummies out of k categorical levels.
<br>
dtype: Data type for new columns. Only a single dtype is allowed. Default value is np.uint8.
<br>
Returns: Dataframe (Dummy-coded data)

### Building a Library Management System [10%]

You are tasked with building a library management system using Python classes and objects. The system should allow for the management of books, library members, and borrowing transactions. Here are the requirements:

1. <b>Create a class called Book with the following attributes and methods: [2%] </b>

    `Attributes:` title, author, isbn, copies_available, total_copies

    `Method:` display_info() that displays information about the book.
    

2. <b>Create a class called Member with the following attributes and methods: [3%] </b>

    `Attributes:` name, member_id, books_borrowed (a list of books borrowed)

    `Method:` borrow_book(book) that allows a member to borrow a book.

    `Method:` return_book(book) that allows a member to return a book.
    

3. <b>Create a class called Library with the following attributes and methods: [5%] </b>

    `Attributes:` name, books (a list of books), members (a list of members)

    `Method:` add_book(book) to add a book to the library.

    `Method:` add_member(member) to add a member to the library.

    `Method:` get_available_books() to get a list of books available for borrowing.

    `Method:` get_borrowed_books() to get a list of borrowed books.

    `Method:` get_member_books(member) to get a list of books borrowed by a member.

Your goal is to create the above class hierarchy and implement the methods to demonstrate interactions between books, members, and the library. 

Test your classes by creating instances, adding books and members, and simulating borrowing and returning transactions.

##### Example outputs:
    John Smith has borrowed 'The Great Gatsby'.
    Alice Johnson has borrowed 'The Great Gatsby'.

    John Smith has returned 'The Great Gatsby'.

```python
class Book:
    def __init__(self,title,author,isbn,copies_available,total_copies):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.copies_available = copies_available
        self.total_copies = total_copies
    
    def display_info(self):
        return f"Book Title is '{self.title}'\nAuthor: {self.author}\nISBN: {self.isbn}\nAvailable copies: {self.copies_available}\nTotal copies:{self.total_copies}"
    
    

book1 = Book("The Great Gatsby","F. Scott Fitzgerald",9780743273565,2,3)
print(book1.display_info())
```

```python
class Member:
    def __init__(self,name,member_id):
        self.name = name
        self.member_id = member_id
        self.books_borrowed = []

        
# create a method to borrow book
    def borrow_book(self,book):
        if book.copies_available > 0:
            book.copies_available -= 1
            self.books_borrowed.append(book)
            return f"{self.name} has borrowed '{book.title}.'"
        else:
            return f"The book '{book.title}' is not available for borrowing"
    
    def return_book(self,book):
        if book.title in self.books_borrowed:
            self.books_borrowed.remove(book.title)
            book.copies_available += 1
            return f"{self.name} has returned '{book.title}.'"
        else:
            return f"The book '{book.title}' was not borrowed."
        
    
            

book1 = Book("The Great Gatsby","F. Scott Fitzgerald",9780743273565,2,3)
member1 = Member('John Smith', 1234)

# John Smith borrow book 1
print(member1.borrow_book(book1))

# add member 2 info
member2 = Member('Alice Johnson',2134)
# Alice Johnson borrowed book 1
print(member1.borrow_book(book1))


# add a new book info, book 2
book2 = Book("To Kill a Mockingbird","Harper Lee",9780060935467,1,4)

#john smith borrowed book 2
print(member1.borrow_book(book2))

#John smith returned book 1
print(member1.return_book(book1))

#print the available copies list
print()
```

```python
class Library:
    def __init__(self,name):
        self.name = name
        self.books = []
        self.members = []
        
#function to add book to the library  
    def add_book(self,book):
        self.books.append(book)
        
#function to add book to the library  
    def add_member(self,member):
        self.members.append(member)

#function to get list of books available for borrowing  
    def get_available_books(self):
        books_available_info = []
        for book in self.books:
            if book.copies_available > 0:
                books_available_info.append(book.title)
           
        return books_available_info

#function to get list of books available for borrowing  
    def get_borrowed_books(self):
        borrowed_count = {}
        for member in self.members:
            for book in member.books_borrowed:
                title = book.title
                if title in borrowed_count:
                    borrowed_count[title] += 1
                else:
                    borrowed_count[title] = 1
        return borrowed_count
    
#function to specific memebr and books borrowed  
    def get_member_books(self, member):
        return [book.title for book in member.books_borrowed]
       

# add books
library = Library("Sengkang Library")
book1 = Book("The Great Gatsby","F. Scott Fitzgerald",9780743273565,2,3)
book2 = Book("To Kill a Mockingbird","Harper Lee",9780060935467,1,4)
library.add_book(book1)
library.add_book(book2)

# add members
member1 = Member('John Smith', 1234)
member2 = Member('Alice Johnson',2134)
library.add_member(member1)
library.add_member(member2)

# borrowing books
print(member2.borrow_book(book1))

# Getting available books
available_books = library.get_available_books()
print("Available Books:", available_books)

# Getting borrowed books
borrowed_books = library.get_borrowed_books()
print("Borrowed Books:", list(borrowed_books.keys()))


# Getting borrowed books by the member
member_to_check = member2
books_borrowed = library.get_member_books(member2)
print(f"Books borrowed by {member_to_check.name}: {books_borrowed}")
```
